Przyjrzyjmy się kątowi wektora początkowego i wynikowego. Oczywiście kąt początkowy to 45st. Co w przypadku wektora wynikowego? Aby olbiczyć jego kąt skorzystamy ze wzoru na tangens kąta, gdyż on wykorzystuje dokładne dwie informacje, które już posiadamy: współrzędnę x i y.

```python
import matplotlib.pyplot as plt
import math
from matplotlib.patches import Arc
```

```python
v = np.array([1, np.sqrt(3)])

# Obliczenie długości wektora
length = np.linalg.norm(v)

# Obliczenie kąta nachylenia w radianach i stopniach
theta_rad = np.arctan2(v[1], v[0])
theta_deg = np.degrees(theta_rad)

# Tworzenie wykresu
fig, ax = plt.subplots()

# Rysowanie wektora
ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Wektor v')

# Rysowanie osi
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Ustawienie zakresu osi
max_val = max(v[0], v[1]) + 1
ax.set_xlim(-1, max_val)
ax.set_ylim(-1, max_val)
# Rysowanie przyprostokątnych
# Przyprostokątna na osi X
ax.plot([0, v[0]], [0, 0], color='red', linestyle='--', linewidth=1)
# Przyprostokątna równoległa do osi Y
ax.plot([v[0], v[0]], [0, v[1]], color='green', linestyle='--', linewidth=1)
# Rysowanie linii reprezentującej kąt

ax.plot([0, v[0]], [0, v[1]], color='green')  # Wektor v

# Rysowanie łuku kąta
arc_radius = 0.5
arc = Arc((0, 0), 2*arc_radius, 2*arc_radius, angle=0, theta1=0, theta2=theta_deg, color='purple')
ax.add_patch(arc)

# Oznaczenie kąta
# Obliczenie pozycji tekstu na podstawie kąta
text_x = arc_radius * np.cos(np.radians(theta_deg / 2)) * 1.1
text_y = arc_radius * np.sin(np.radians(theta_deg / 2)) * 1.1
ax.text(text_x, text_y, f'θ', color='purple', fontsize=12)

# Oznaczenie osi
ax.text(max_val, -0.5, 'X', fontsize=12, ha='center')
ax.text(-0.5, max_val, 'Y', fontsize=12, ha='center')

# Dodanie siatki
ax.grid(True, which='both')

# Ustawienie proporcji osi
ax.set_aspect('equal')

# Dodanie legendy
ax.legend()

# Wyświetlenie wykresu
plt.title('Wektor v = (1, √3) oraz jego kąt nachylenia')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.show()

# Wyświetlenie wartości kąta
print(f'Kąt nachylenia wektora v względem osi X wynosi: {theta_deg:.2f}°')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Arc

# Definicja wektorów
A = np.array([1, 1])
B = np.array([1, np.sqrt(3)])

# Obliczenie długości wektorów
length_A = np.linalg.norm(A)
length_B = np.linalg.norm(B)

# Obliczenie kątów nachylenia w radianach i stopniach
theta_A_rad = np.arctan2(A[1], A[0])
theta_A_deg = np.degrees(theta_A_rad)

theta_B_rad = np.arctan2(B[1], B[0])
theta_B_deg = np.degrees(theta_B_rad)

# Tworzenie wykresu
fig, ax = plt.subplots(figsize=(10, 10))

# Rysowanie wektorów
ax.quiver(0, 0, A[0], A[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Wektor A (1, 1)')
ax.quiver(0, 0, B[0], B[1], angles='xy', scale_units='xy', scale=1, color='orange', label='Wektor B (1, √3)')

# Rysowanie osi
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

# Ustawienie zakresu osi
max_val = max(np.linalg.norm(A), np.linalg.norm(B)) + 1
ax.set_xlim(-1, max_val)
ax.set_ylim(-1, max_val)

# Rysowanie przyprostokątnych dla wektora A
# Przyprostokątna na osi X
ax.plot([0, A[0]], [0, 0], color='blue', linestyle='--', linewidth=1)
# Przyprostokątna równoległa do osi Y
ax.plot([A[0], A[0]], [0, A[1]], color='blue', linestyle='--', linewidth=1)

# Oznaczenie przyprostokątnych dla wektora A
# ax.text(A[0]/2, -0.3, 'Przyprostokątna X', color='blue', fontsize=12, ha='center')
# ax.text(A[0] + 0.3, A[1]/2, 'Przyprostokątna Y', color='blue', fontsize=12, va='center')

# Rysowanie przyprostokątnych dla wektora B
# Przyprostokątna na osi X
ax.plot([0, B[0]], [0, 0], color='orange', linestyle='--', linewidth=1)
# Przyprostokątna równoległa do osi Y
ax.plot([B[0], B[0]], [0, B[1]], color='orange', linestyle='--', linewidth=1)

# Oznaczenie przyprostokątnych dla wektora B
# ax.text(B[0]/2, -0.3, 'Przyprostokątna X', color='orange', fontsize=12, ha='center')
# ax.text(B[0] + 0.3, B[1]/2, 'Przyprostokątna Y', color='orange', fontsize=12, va='center')

# Rysowanie łuków kąta dla wektorów A i B
arc_radius = 0.5

# Łuk dla wektora A
arc_A = Arc((0, 0), 2*arc_radius, 2*arc_radius, angle=0, theta1=0, theta2=theta_A_deg, color='purple', linewidth=2)
ax.add_patch(arc_A)

# Oznaczenie kąta dla wektora A
text_x_A = arc_radius * np.cos(np.radians(theta_A_deg / 2)) * 1.1
text_y_A = arc_radius * np.sin(np.radians(theta_A_deg / 2)) * 1.1
ax.text(text_x_A, text_y_A, f'θ_A', color='purple', fontsize=12)

arc_radius = 0.7
# Łuk dla wektora B
arc_B = Arc((0, 0), 2*arc_radius, 2*arc_radius, angle=0, theta1=0, theta2=theta_B_deg, color='green', linewidth=2)
ax.add_patch(arc_B)


# Oznaczenie kąta dla wektora B
text_x_B = arc_radius * np.cos(np.radians(theta_B_deg / 2)) * 1.1
text_y_B = arc_radius * np.sin(np.radians(theta_B_deg / 2)) * 1.1
ax.text(text_x_B, text_y_B, f'θ_B', color='green', fontsize=12)

# Rysowanie przekształcenia wektora A w wektor B
# Opcjonalnie: strzałka pokazująca przekształcenie
# ax.annotate('', xy=(B[0], B[1]), xytext=(A[0], A[1]),
#             arrowprops=dict(facecolor='red', shrink=0.05, width=2, headwidth=8),
#             label='Przekształcenie A → B')

# Oznaczenie osi
ax.text(max_val, -0.5, 'X', fontsize=14, ha='center')
ax.text(-0.5, max_val, 'Y', fontsize=14, ha='center')

# Dodanie siatki
ax.grid(True, which='both', linestyle=':', linewidth=0.5)

# Ustawienie proporcji osi
ax.set_aspect('equal')

# Dodanie legendy
ax.legend(loc='upper left')

# Wyświetlenie wykresu
plt.title('Przekształcenie wektora A = (1, 1) w wektor B = (1, √3) oraz ich kąty nachylenia')
plt.xlabel('Oś X')
plt.ylabel('Oś Y')
plt.show()

# Wyświetlenie wartości kątów w konsoli
print(f'Kąt nachylenia wektora A względem osi X wynosi: {theta_A_deg:.2f}°')
print(f'Kąt nachylenia wektora B względem osi X wynosi: {theta_B_deg:.2f}°')
```
